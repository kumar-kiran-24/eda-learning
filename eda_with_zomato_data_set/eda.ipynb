```python
e_path=os.path.join("Zomatodataset","Country-Code.xlsx")
df_country=pd.read_excel(e_path)
df_country.head()
```

combine data set

```python
final_df=pd.merge(df,df_country,on="Country Code",how="left")
final_df.head(2)
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
country_values=final_df.Country.value_counts().values
```

create the pie chart

```python
#top 3 countries that use zomato
plt.pie(x=country_values[:3],labels=country_names[:3],autopct="%1.2f%%")
plt.show()
```

obeservation: zomato maximumum records  are from india after that usa and uk

```python
ratings=final_df.groupby(["Aggregate rating",'Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
ratings
```
